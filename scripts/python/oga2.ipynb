```python
from openeye.oechem import *
import os,sys
directory = "/Users/jfeng1/Datasets/OGA"
files = ["output1.sdf","output2.sdf","output3.sdf"]
smiDict = {}
for f in files:
    f = os.path.join(directory,f)
    ifs = oemolistream()
    print ifs.open(f)
    mol = OEGraphMol()
    while OEReadMolecule(ifs,mol):
        smiles = OEMolToSmiles(mol)
        smiDict[smiles] = OEGetSDData(mol,"plate_ID")
    ifs.close()
print len(smiDict)
```

```python
file = os.path.join(directory,"")
ifs = oemolistream()
ifs.open(file)
mol = OEGraphMol()
n = 0
while OEReadMolecule(ifs,mol):
    smiles = OEMolToSmiles(mol)
    if smiDict.has_key(smiles):
        n += 1
print n
```

```python
file = os.path.join(directory,"lead_like_plates.txt")
lines = open(file,"r").read().splitlines()
plates = []
for line in lines:
    plates.append(line.split()[0])
n = 0
for v in smiDict.values():
    if v in plates:
        n += 1
print n
```

```python
from pylab import *
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['text.color']='black'

# make the pie circular by setting the aspect ratio to 1
plt.figure(figsize=(10,10))
values = [832,230,181,477]
labels = ['Biogen-HTS (180K)', 'Plate-Based Selection (Lead-like) (100K)', 'Plate-based Selection (ROCS) (70K)', 'Other (CRL etc.) (150K)']

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.0f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct

patches, texts, autotexts = plt.pie(values, labels=labels, colors=['y','g','r','cyan'],autopct=make_autopct(values))
for t in autotexts:
    t.set_fontsize(15)
for t in texts:
    t.set_fontsize(15)

plt.show()
```
