```python
from openeye.oechem import *
from openeye.oegraphsim import *
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.manifold import MDS

def tanimoto_dist(a,b):
    dotprod = np.dot(a,b)
    tc = dotprod / (np.sum(a) + np.sum(b) - dotprod)
    return 1.0-tc

sdf = "data.sdf"

ifs = oemolistream()
ifs.open(sdf)
oemol = OEGraphMol()

fpList = []
idx = 0
while OEReadMolecule(ifs,oemol):
    
    fp = OEFingerPrint()
    OEMakeFP(fp,oemol,OEFPType_Circular)
    desc = []
    for i in range(0,fp.GetSize()):
        if fp.IsBitOn(i):
            desc.append(1)
        else:
            desc.append(0)
    fpList.append(np.array(desc))
    idx+=1
ifs.close()

```

```python
mds = MDS(n_components=4096, metric=tanimoto_dist)
data = mds.fit_transform(fpList)
```

```python
from matplotlib import pyplot as plt
plt.scatter(data[:,0],data[:,1])
plt.show()
```
