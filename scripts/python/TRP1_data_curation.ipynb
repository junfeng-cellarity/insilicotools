```python
from openeye.oechem import *
directory = "/Users/jfeng1/forRab"
import os,glob
```

```python
files = glob.glob(os.path.join(directory,"*.sdf"))
print files
```

```python
molecules = {}
mol_names = []
for f in files:
    ifs = oemolistream()
    ifs.open(f)
    mol = OEGraphMol()
    while OEReadMolecule(ifs,mol):
        mol_name = OEGetSDData(mol,"ID")
        if mol_name not in mol_names:
            mol_names.append(mol_name)
            mol.SetTitle(mol_name)
            OEClearSDData(mol)
            molecules[mol_name] = OEGraphMol(mol)
    ifs.close()

print mol_names,len(mol_names)
print molecules,len(molecules)
data_dict = {}
for mol_name in mol_names:
    data_dict[mol_name] = {}

secondary_columns = ["PH","SPECIES"]
for f in files:
    ifs = oemolistream()
    ifs.open(f)
    mol = OEGraphMol()
    while OEReadMolecule(ifs,mol):
        mol_name = OEGetSDData(mol,"ID")
        secondary_tag = ""
        for dp in OEGetSDDataPairs(mol):
            tag = dp.GetTag()
            value = dp.GetValue()
            if tag == "ID":
                continue
            elif tag in secondary_columns:
                secondary_tag = tag
            else:
                if secondary_tag is not None:
                    tag = tag+OEGetSDData(mol,secondary_tag)
                    data_dict[mol_name][tag] = value
    ifs.close()

ofs = oemolostream()
ofs.open(os.path.join(directory,"output.sdf"))
for mol_name in mol_names:
    mol = molecules[mol_name]
    for tag in data_dict[mol_name].keys():
        OESetSDData(mol,tag,data_dict[mol_name][tag])
    OEWriteMolecule(ofs,mol)
ofs.close()
    
    
    
    
```
